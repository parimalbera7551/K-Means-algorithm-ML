```python
import numpy as np
import random
from PIL import Image
import matplotlib.pyplot as plt

def load_image(path):
    img = Image.open(path)
    img_array = np.array(img)
    return img_array

def initialize_centroids(data, k):
    centroids_idx = random.sample(range(data.shape[0]), k)
    centroids = data[centroids_idx]
    return centroids

def assign_clusters(data, centroids):
    distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
    # Assign each data point to the nearest centroid
    clusters = np.argmin(distances, axis=1)
    return clusters

def update_centroids(data, clusters, k):
    centroids = np.zeros((k, data.shape[1]))
    for i in range(k):
        centroids[i] = np.mean(data[clusters == i], axis=0)
    return centroids

def k_means(data, k, max_iters=100):
    centroids = initialize_centroids(data, k)
    
    for _ in range(max_iters):
        clusters = assign_clusters(data, centroids)
        new_centroids = update_centroids(data, clusters, k)
        if np.allclose(centroids, new_centroids):
            break
        
        centroids = new_centroids
    
    return clusters

# Load your 4-band images
image_paths = ['band1.gif', 'band2.gif', 'band3.gif', 'band4.gif']
images = [load_image(path) for path in image_paths]

# Flatten each image
flattened_images = [img.flatten()[:, np.newaxis] for img in images]

# Concatenate the flattened images
combined_data = np.concatenate(flattened_images, axis=1)

# Values of k
k_values = [2, 3, 4, 5, 6]

# Dictionary to store cluster labels for each k value
cluster_labels = {}

# Perform K-means clustering for each value of k
for k in k_values:
    clusters = k_means(combined_data, k)
    cluster_labels[k] = clusters

# Print the shape of the labels for each k value
for k, labels in cluster_labels.items():
    print(f"Shape of cluster labels for k={k}: {labels.shape}")

# Reshape the clusters array to have the same shape as the original images
num_images = len(images)
for k in k_values:
    # Reshape clusters to match the shape of the original images
    clusters = cluster_labels[k].reshape(512, 512)
    print(clusters)
```

```python

random.seed(42)

# Visualize the clustered images for each k value
fig, axs = plt.subplots(1, 5, figsize=(20, 5))
for i, k in enumerate(k_values):
    # Reshape clusters to match the shape of the original images
    clusters = cluster_labels[k].reshape(-1, 512, 512)  # Reshape to original image dimensions
    
    # Choose a random index to select a random image from the clustered images
    random_index = np.random.randint(clusters.shape[0])  # Use clusters.shape[0] as num_images
    
    # Select the random image
    random_image = clusters[random_index]
    
    # Plot the clustered image
    axs[i].imshow(random_image, cmap='viridis')
    axs[i].set_title(f'K-means Clustering with k={k}')
    axs[i].axis('off')

plt.tight_layout()
plt.show()
```
